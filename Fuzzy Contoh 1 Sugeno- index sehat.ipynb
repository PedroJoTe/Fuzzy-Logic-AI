```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline

# Generate universe variables
x_tinggi = np.arange(100, 201, 1)
x_berat = np.arange(30, 101, 1)

# Generate fuzzy membership functions
tinggi_sp = fuzz.trapmf(x_tinggi, [0, 0, 115, 120])
tinggi_p  = fuzz.trapmf(x_tinggi, [115, 120, 140, 145])
tinggi_s  = fuzz.trapmf(x_tinggi, [140, 145, 160, 165])
tinggi_t  = fuzz.trapmf(x_tinggi, [160, 165, 180, 185])
tinggi_st = fuzz.trapmf(x_tinggi, [180, 185, 201, 201])

berat_sk = fuzz.trapmf(x_berat, [0, 0, 40, 45])
berat_k  = fuzz.trapmf(x_berat, [40, 45, 50, 55])
berat_m  = fuzz.trapmf(x_berat, [50, 55, 60, 65])
berat_b  = fuzz.trapmf(x_berat, [60, 65, 80, 85])
berat_sb = fuzz.trapmf(x_berat, [80, 85, 101, 101])

SS  =0.8
S   =0.6
AS  =0.4
TS  =0.2

# Visualize these universes and membership functions
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 8))

ax0.plot(x_tinggi, tinggi_sp, 'b', linewidth=1.5, label='Sangat Pendek')
ax0.plot(x_tinggi, tinggi_p, 'g', linewidth=1.5, label='Pendek')
ax0.plot(x_tinggi, tinggi_s, 'r', linewidth=1.5, label='Sedang')
ax0.plot(x_tinggi, tinggi_t, 'y', linewidth=1.5, label='Tinggi')
ax0.plot(x_tinggi, tinggi_st, 'grey', linewidth=1.5, label='Sangat Tinggi')

ax0.set_title('Tinggi Badan')
ax0.legend()

ax1.plot(x_berat, berat_sk, 'b', linewidth=1.5, label='SAngat Kurus')
ax1.plot(x_berat, berat_k, 'g', linewidth=1.5, label='Kurus')
ax1.plot(x_berat, berat_m, 'r', linewidth=1.5, label='Biasa')
ax1.plot(x_berat, berat_b, 'y', linewidth=1.5, label='Berat')
ax1.plot(x_berat, berat_sb, 'grey', linewidth=1.5, label='Sangat BErat')

ax1.set_title('Berat BAdan')
ax1.legend()

ax2.plot([4, 4], [0, SS], 'b', linewidth=1.5, label='Sangat Sehat')
ax2.plot([3, 3], [0, S], 'g', linewidth=1.5, label='Sehat')
ax2.plot([2, 2], [0, AS], 'r', linewidth=1.5, label='Agak Sehat')
ax2.plot([1, 1], [0, TS], 'y', linewidth=1.5, label='Kurang Sehat')

ax2.set_title('Kesehatan')
ax2.legend()
# Turn off top/right axes
for ax in (ax0, ax1):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
## input Crisp
in_tinggi = 170
in_berat = 54

alamat="/Users/hellemanembu/Pinrolinvic/Desktop/Kuliah 2018/Sistem Cerdas/images/TB.png"
Image(filename=alamat, width=500, height=400)
```

```python

# Fuzzification
in_1 = []
in_1.append(fuzz.interp_membership(x_tinggi, tinggi_sp, in_tinggi))
in_1.append(fuzz.interp_membership(x_tinggi, tinggi_p, in_tinggi))
in_1.append(fuzz.interp_membership(x_tinggi, tinggi_s, in_tinggi))
in_1.append(fuzz.interp_membership(x_tinggi, tinggi_t, in_tinggi))
in_1.append(fuzz.interp_membership(x_tinggi, tinggi_st, in_tinggi))

in_2 = []
in_2.append(fuzz.interp_membership(x_berat, berat_sk, in_berat))
in_2.append(fuzz.interp_membership(x_berat, berat_k, in_berat))
in_2.append(fuzz.interp_membership(x_berat, berat_m, in_berat))
in_2.append(fuzz.interp_membership(x_berat, berat_b, in_berat))
in_2.append(fuzz.interp_membership(x_berat, berat_sb, in_berat))

print("Derajat Keanggotaan Tinggi-Badan:")
if in_1[0]>0 :
    print("Sangat Pendek :"+ str(in_1[0]))
if in_1[1]>0 :
    print("Pendek        :"+ str(in_1[1]))
if in_1[2]>0 :
    print("Sedang        :"+ str(in_1[2]))
if in_1[3]>0 :
    print("Tinggi        :"+ str(in_1[3]))
if in_1[4]>0 :
    print("Sangat tinggi :"+ str(in_1[4]))

print("")
print("Derajat Keanggotaan Berat-Badan:")
if in_2[0]>0 :
    print("Sangat Kurus  :"+ str(in_2[0]))
if in_2[1]>0 :
    print("Kurus         :"+ str(in_2[1]))
if in_2[2]>0 :
    print("biasa         :"+ str(in_2[2]))
if in_2[3]>0 :
    print("Berat         :"+ str(in_2[3]))
if in_2[4]>0 :
    print("Sangat berat  :"+ str(in_2[4]))
```

```python
print("Matriks Tinggi-Badan:")
print(in_1)
print("")
print("Matriks Berat-Badan:")

print(in_2)
```

```python
## Inference dan Defazzification dengan Centroid Sugeno: 
#Penyebut
rul = []
for i in range(5):
    for j in range(5):
        rule=fuzz.relation_min(in_1[i], in_2[j])
        rul.append(rule)
        #print(rule)
penyebut=np.sum(rul)
#Pembilang
rul = []
for i in range(5):
    for j in range(5):
        rule=fuzz.relation_min(in_1[i], in_2[j])
        rulxx= rule*FAM[i][j]
        rul.append(rulxx)
        #print(rulxx)
pembilang=np.sum(rul)
hasil = pembilang/penyebut

## Logic untuk hasil
print("Index Kesehatan :"+ str(hasil))
if hasil >=0.2 and hasil < 0.4 :
    za = (abs(hasil-0.2)/(0.4-0.2))*100
    zb = (abs(hasil-0.4)/(0.4-0.2))*100
    print("Tidak Sehat :"+ '{:2.2f}'.format(zb)+" %")
    print("Agak Sehat  :"+ '{:2.2f}'.format(za)+" %")
if hasil>=0.4 and hasil< 0.6:
    za = (abs(hasil-0.4)/(0.6-0.4))*100
    zb = (abs(hasil-0.6)/(0.6-0.4))*100
    print("Agak Sehat :"+ '{:2.2f}'.format(zb)+" %")
    print("Sehat      :"+ '{:2.2f}'.format(za)+" %")
if hasil>=0.6 and hasil< 0.8:
    za = (abs(hasil-0.6)/(0.8-0.6))*100
    zb = (abs(hasil-0.8)/(0.8-0.6))*100
    print("Sehat        :"+ '{:2.2f}'.format(zb)+" %")
    print("Sangat Sehat :"+ '{:2.2f}'.format(za)+" %")
if hasil>=0.8:
    za = (abs(hasil-0.8)/(1.0-0.8))*100
    print("Sangat Sehat :"+ '{:2.2f}'.format(za)+" %")

```
